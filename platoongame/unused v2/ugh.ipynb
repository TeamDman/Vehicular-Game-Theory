```python
from typing import List, Tuple
from sumtree import SumTree
import torch
import random
batch_size = 10

tree = SumTree(100)
for i in range(100):
    tree.add(random.random()*4)

print(tree.get(25.3))

ranges = torch.linspace(0, tree.total, batch_size+1).int().tolist()
ranges = torch.as_tensor(list(zip(ranges[:-1], ranges[1:])))
print(ranges)
rand = torch.rand(batch_size)
print(rand)
cumsums = ranges[:,0] + (ranges[:,1]-ranges[:,0])*rand
print(cumsums)
items = [tree.get(x.item()) for x in cumsums]
print(items)
print("uwu")
tree.update(0,10)
print(tree.get(0))
```
