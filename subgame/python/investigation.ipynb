```python
import json

with open("human.json","r") as f:
    human = json.load(f)

with open("ai.json","r") as f:
    ai = json.load(f)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({"figure.figsize":(7,5), "figure.dpi":100})

bins = np.linspace(0,60,60)
plt.hist(human, bins, alpha=0.5, label="human")
plt.hist(ai, bins, alpha=0.5, label="ai")
plt.legend(loc="upper right")
plt.title("Utility using actions chosen by human and ai strategies")
plt.show()
# plt.gca().set(title="Freq histogram", ylabel="Freq")
```

```python
from random import randint
hscore = 0
ascore = 0
for i in range(10000000):
    hscore += human[randint(0,len(human)-1)]
    ascore += ai[randint(0,len(ai)-1)]
print(hscore)
print(ascore)
```
